# Research Question

## Background and Prior Work

According to the Cleveland Clinic, thyroid gland is a part of endocrine system and located around the base of the Larynx (voice box), surrounding the trachea (windpipe) and is responsible for creating necessary hormones to regulate the body functions like metabolism through chemical reaction and control hormones by releasing the hormones into the bloodstream. In fact, all cells require energy from metabolized food in order to function so thyroid plays an important role in maintaining that our body is working properly. It is a fairly common disease as stated by Cleveland Clinic, diseases that degrade or over activate the thyroid function can have a major impact on our body functionality such as fatigue, weight loss/gain, depression, heavy or irregular menstrual periods, mood swing, etc… However, the diseases are not debilitating and are all treatable through the use of medication, therapy, or surgery.

The most common diagnoses of thyroid diseases are hypothyroidism and hyperthyroidism where the patients do not have the adequate amount of the thyroid hormones. This disease has a huge margin of impact on the population as it is prevalent in roughly around 12 percent of the U.S. population and around 20 million people exhibit the thyroid disease’s symptoms. By using the a large dataset as a way to implement machine learning and data analysis to make an inference on early detection and the effectiveness of treatment can lead to minimizing the proportion of population that are impacted by thyroid disease.

# Model 1

Drop all yes/no columns related to lab work and results from lab work columns since they contains many nan values.   

```python
df['target'] = np.where(df['target'] == '-', 0, 1)
df['target'].value_counts()
```

```python
df1 = df.loc[:, ['age', 'sex', 'on_thyroxine', 'query_on_thyroxine',
       'on_antithyroid_meds', 'sick', 'pregnant', 'thyroid_surgery',
       'I131_treatment', 'query_hypothyroid', 'query_hyperthyroid', 'lithium',
       'goitre', 'tumor', 'hypopituitary', 'psych', 'target']]

categorical_vars = ['sex', 'on_thyroxine', 'query_on_thyroxine', 'on_antithyroid_meds', 'sick', 
                    'pregnant', 'thyroid_surgery', 'I131_treatment', 'query_hypothyroid', 
                    'query_hyperthyroid', 'lithium', 'goitre', 'tumor', 'hypopituitary', 'psych','target']

df1[categorical_vars] = df1[categorical_vars].astype('category')
df1['target'] = df1['target'].cat.codes

response = 'target'
covariates1 = df1.columns.drop(response)
formula1 = f"{response} ~ " + " + ".join(covariates1)

model1 = smf.logit(formula1, data=df1).fit()
print(model1.summary())
```

```python
significant_covariates_and_pvalues = model1.pvalues[model1.pvalues < 0.15]
print(significant_covariates_and_pvalues)
```

```python
formula1_significant = f"{response} ~ " + 'sex + on_thyroxine + pregnant + query_hypothyroid + query_hyperthyroid + lithium + goitre + psych'

model1_significant = smf.logit(formula1_significant, data=df1).fit()
model1.pvalues<0.15
```
