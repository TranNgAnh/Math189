# Feature selections

We are going to try backward selection, forward selection and LASSO penalty as a way to see which variables of the blood test are best fit the linear regression model. Due to the fact that most of this data are categorical, it is not feasible to apply the selected variables to the linear regression model because it will not give us predicting values that are probability. Only the continuous data can be fit over the linear regression model.

These functions are provided by Professor Siddharth Vishwanath in Spring 2024, MATH 189.

```python
def add(df, selected_columns, columns, criterion, response='y'):
    best_criterion = np.inf
    best_column = None
    for column in columns - selected_columns:
        new_columns = selected_columns.union({column})
        formula = f'{response} ~ {" + ".join(new_columns)}'
        current_criterion = criterion(formula, df)
        if current_criterion < best_criterion:
            best_criterion = current_criterion
            best_column = column
    return selected_columns.union({best_column}), best_criterion
def forward(df, criterion, response='y'):
    selected_columns = set()
    columns = set(df.columns.drop(response))
    best_criterion = np.inf
    while len(selected_columns) < len(columns):
        potential_columns, current_criterion = add(df, selected_columns, columns, criterion, response)
        if current_criterion > best_criterion:
            break
        else:
            selected_columns = potential_columns
            best_criterion = current_criterion
            print(f'Criterion: {best_criterion}')
    return selected_columns
def remove(df, selected_columns, criterion, response='y'):
    worst_criterion = np.inf
    worst_column = None
    for column in selected_columns:
        new_columns = selected_columns - {column}
        formula = f'{response} ~ {" + ".join(new_columns)}'
        current_criterion = criterion(formula, df)
        if current_criterion < worst_criterion:
            worst_criterion = current_criterion
            worst_column = column
    return worst_column, worst_criterion
def backward(df, criterion, response='y'):
    selected_columns = set(df.columns.drop(response))
    columns = set(df.columns.drop(response))
    best_criterion = np.inf
    while len(selected_columns) > 0:
        potential_column, current_criterion = remove(df, selected_columns, criterion, response)
        if current_criterion > best_criterion:
            break
        else:
            selected_columns = selected_columns - {potential_column}
            best_criterion = current_criterion
            print(f'Criterion: {best_criterion}')
    return selected_columns
```

Given that the thyroid dataset seems to be relatively large (as indicated by the number of unique target values), and potentially includes many features, BIC might be more appropriate due to its stricter penalty on model complexity compared to AIC. This can help in selecting a simpler and more interpretable model, which can be important in medical contexts where model interpretability is crucial.


```python
df_cont = df.loc[:,['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'target']]
response = 'target'

formula = f'{response} ~ {" + ".join(df_cont.columns[:-1])}'
print(formula)
print(df_cont)
```

```python
bic_criterion = lambda formula, df: smf.ols(formula, data=df).fit().bic
```

```python
f_vars_bic = forward(df_cont, bic_criterion, response = response)
print(f_vars_bic)
```

```python
f_formula = f'{response} ~ {" + ".join(f_vars_bic)}'
f_formula 
```

```python
b_vars_bic = backward(df_cont, bic_criterion, response = response)
print(b_vars_bic)
```

```python
b_formula = f'{response} ~ {" + ".join(b_vars_bic)}'
b_formula 
```

LASSO penalty

```python
lambdas = np.linspace(0.01, 1, 101)
elastic_coefs = []

for lam in lambdas:
    elastic_fit = smf.ols(formula, data=df_cont).\
                fit_regularized(method='elastic_net', alpha=lam, L1_wt=1.0)
    elastic_coefs.append(elastic_fit.params[1:])

elastic_coefs = pd.DataFrame(elastic_coefs, index=lambdas)

fig, ax = plt.subplots(figsize=(7, 7))
for column in elastic_coefs.columns:
    ax.plot(lambdas, elastic_coefs[column], label=column)
ax.set_xlabel('lambda')
ax.set_ylabel('coefficient')
```

```python
def find_lambda_zero(coefs):
    zero_points = {}
    for col in coefs.columns:
        zero_lambda = lambdas[np.where(coefs[col].values == 0)[0]]
        if zero_lambda.size > 0:
            zero_points[col] = zero_lambda[0]
    return zero_points
lambda_zeros = find_lambda_zero(elastic_coefs)
lambda_zeros
```

```python
elastic_fit = smf.ols(formula, data=df_cont).\
                fit_regularized(method='elastic_net', alpha=0.01, L1_wt=1.0)
print(elastic_fit.params[elastic_fit.params != 0])
```

```python
lasso_vars = elastic_fit.params[elastic_fit.params != 0].index.tolist()
lasso_formula = f'{response} ~ {" + ".join(lasso_vars)}'
lasso_formula 
```

As we can see, the variables that got returned by forward selection, backward selection using BIC as a criterion are the same and LASSO penalty are completely different. We will be testing each of these formula to see which would give us the best result and interpretation.
